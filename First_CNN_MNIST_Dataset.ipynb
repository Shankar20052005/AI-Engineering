```python
cnn = SimpleCNN()  # create model
dataiter = iter(trainloader)  # get 1 image
images, labels = next(dataiter)

# Run the image through the CNN
_ = cnn(images)

print("Image shape:", images.shape)
print("Feature map 1 shape:", cnn.feat1.shape)
print("Feature map 2 shape:", cnn.feat2.shape)
```

```python
# Show input image
plt.imshow(images[0][0].detach(), cmap='gray')
plt.title("Input Image (Digit: {})".format(labels.item()))
plt.axis('off')
plt.show()

# Show all 4 feature maps from conv1
for i in range(4):
    plt.imshow(cnn.feat1[0][i].detach(), cmap='gray')
    plt.title(f"Feature Map {i+1} from Conv1")
    plt.axis('off')
    plt.show()
```
